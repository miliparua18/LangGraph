```python
from langgraph.store.memory import InMemoryStore
```

```python
#creating a store
store = InMemoryStore()
```

```python
#creating a namespace
namespace = ("user", "u1")
```

# Creating Memory

```python
store.put(namespace, "1", {"data": "user likes pizza"})
store.put(namespace, "2", {"data": "user prefers dark mode"})
```

```python
namespace2 = ("user", "u2")

store.put(namespace2, "1", {"data": "user likes pasta"})
store.put(namespace2,"2",{"data": "user prefers grid style nevigation"})
```

# Retrieving Memory


```python
store.get(namespace, "1")
```

# Retrieving all memory

```python
items = store.search(namespace2)

for item in items:
    print(item.value)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
store = InMemoryStore(index={'embed': embedding_model, 'dims': 1536})
```

```python
namespace = ("user", "u1")
```

```python
store.put(namespace, "1", {'data': 'user prefer concise answer'})
store.put(namespace, "2", {"data": "user likes simple explanations"})
store.put(namespace, "3", {"data": "user is learning LangGraph"})
store.put(namespace, "4", {"data": "user is interested in AI and ML"})
store.put(namespace, "5", {"data": "user prefers Python examples"})
store.put(namespace, "6", {"data": "user is working on RAG applications"})
store.put(namespace, "7", {"data": "user likes beginner-friendly tutorials"})
store.put(namespace, "8", {"data": "user is learning vector databases"})
store.put(namespace, "9", {"data": "user is building chatbot projects"})
store.put(namespace, "10", {"data": "user prefers step-by-step guidance"})
store.put(namespace, "11", {"data": "user is exploring long term memory systems"})
```

```python
items = store.search(namespace, query="what is the user building", limit = 1)
for item in items:
    print(item.value)
```
